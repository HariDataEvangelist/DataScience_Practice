```python
def f():
    yield True

try:
    g = f()
    h = next(g)
    assert h is True
    print("True")
    h=next(g)
    assert h is None
    print("None")
except AssertionError:
    print("Assertion failed")
except GeneratorExit:
    print("Exit")
except StopIteration:
    print("Stop")
```
